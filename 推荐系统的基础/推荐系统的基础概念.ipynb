# 推荐系统的意义
+ 用户角度：解决在“信息过载”的情况下，用户如何高效的获取感兴趣的信息的问题
+ 公司角度：解决产品能够最大化的吸引用户，留住用户，实现收益的增长。

**这两个角度在一个健康的商业模式下一定是互相成就，和谐统一的！！**
> 比如书中所居的 YouTube，以及日常使用的软件如淘宝、抖音、小红书，推荐系统的作用就是同时满足以上两条。

# 推荐系统的架构

推荐系统解决的是 **人** 和 **海量信息** 之间的关系！
同时会受到**场景信息**的影响！

## 逻辑框架


简单来说就是使用用户信息，物品信息，场景信息，来从海量的物品中选择出用户感兴趣的。

## 技术架构
> 需要考虑的问题：
> + 用户信息，物品信息，场景信息分别是什么，怎么存储，怎么更新，怎么使用？
> + 使用什么模型与算法，如何训练、如何预测、如何优化？

数据部分中有离线的、实时的
模型部分需要训练、评估、部署、线上推断等。

# 实验流程与 A/B测试

算法工程师的任务就是改进相关模型、策略、算法、特征等，然后首先离线测试，之后在将用户分组做对比实验，如果指标显著由于对照组，那么就全流量使用。
## A/B测试
可以帮助查看策略是否真的好，也可以帮助选取模型的最优参数！
对照实验！！！
### 随机分桶
实际上就是指设置多个组别，来对照

### 分层实验
> background：一个公司有很多个算法工程师，他们分别的主要工作在推荐系统的不同链路上，都需要做实验，那样在按照普通的分组就会发现人数不够分，也就是实验材料不够的情况。

# 推荐系统的链路

## 召回

有多个不同的找回通道先从物品库中选出一定数目的物品，之后需要去重、过滤等操作，其中过滤是指排除掉用户不喜欢的话题等策略
## 排序

之后是对召回的物品进行排序。为了降低计算量，将排序分为粗排和精排。
粗排是使用简单的，参数小的模型。
精排是使用更加复杂、精确的、参数更大的模型
排序后带着对物品的打分一起输出。
## 重排

排序的分数一般不直接展示，因为要考虑多样性、推广策略等问题，对展示的物品顺序等重新考量！
> 多样性、避免同质化、生态要求……
